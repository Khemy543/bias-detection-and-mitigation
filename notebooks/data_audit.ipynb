```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('recruitment_dataset.csv')
S = "gender"    # race, age etc (loop later)
y = "hired"
```

```python
def base_rates(df, S, y):
    return df.groupby(S)[y].agg(['mean','count']).rename(columns={'mean':'P(y=1)'})

def disparate_impact(df, S, y, ref):
    rates = df.groupby(S)[y].mean()
    return (rates / rates.loc[ref]).sort_values()
```

```python

print(base_rates(df, S, y))
print(disparate_impact(df, S, y, ref='Male'))
```
