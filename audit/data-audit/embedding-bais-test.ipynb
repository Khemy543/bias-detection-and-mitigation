```python
import pandas as pd
import spacy

# Load dataset
df = pd.read_csv("../../datasets/recruitment.csv")

# Text-based feature to analyze
df['text_features'] = (
    df['Skills'] + ' ' + 
    df['Certifications'] + ' ' + 
    df['Job_Role_Applied']
).fillna('').str.lower()

nlp = spacy.load("en_core_web_md")

# Convert text features to embeddings
df['docs'] = df['text_features'].apply(nlp)

# Access word embeddings
word_vectors = nlp("python")
```
