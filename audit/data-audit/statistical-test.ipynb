```python
# To determine whether differences in selection outcomes (Selected) across demographic groups
# (like Gender, Race, or Age) are statistically significant — meaning that bias is likely systemic, not random.

# Representation imbalance doesn’t always mean unfairness — sometimes differences arise by chance.
# Statistical tests add rigor by measuring whether group differences in hiring outcomes are significant.
# If results show p < 0.05, it means there’s less than a 5% probability that the observed difference happened by chance → evidence of bias.
```

```python
# CHI-Squared Test for Independence
# H0: Selection outcome is independent of demographic group (no bias)
# H1: Selection outcome is dependent on demographic group (potential bias)
# If p < 0.05, reject H0 → evidence of bias.
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load dataset
df = pd.read_csv("../../datasets/recruitment.csv")
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65], labels=['18-25', '26-35', '36-45', '46-55', '56-65'])

# CHI-Squared Test for Independence
def chi_squared_test(df, attribute, target='Selected'):
    contingency_table = pd.crosstab(df[attribute], df[target])
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print(f"Chi-Squared Test for {attribute}:")
    print(contingency_table)
    print(f"Chi2 = {chi2:.3f}, df = {dof}, p-value = {p:.4f}")
    if p < 0.05:
        print(f"→ Statistically significant association between {attribute} and {target} (possible bias)\n")
    else:
        print(f"→ No significant relationship detected.\n")
    return chi2, p


# CHI-Squared Test for Gender
chi_squared_test(df, 'Gender')

# CHI-Squared Test for Age
chi_squared_test(df, 'Age')

# CHI-Squared Test for Race
chi_squared_test(df, 'Race')
```

```python
# T-test / ANOVA for comparing means across groups

# H0: Mean scores are equal across demographic groups (no bias)
# H1: At least one group has a different mean score (potential bias)
# If p < 0.05, reject H0 → evidence of bias.

from scipy.stats import f_oneway, kruskal

# ANOVA Test method
def anova_test(attribute, numeric_feature, non_parametric=False):
    groups = [df.loc[df[attribute] == level, numeric_feature] for level in df[attribute].unique()]

    if non_parametric:
        stat, p = kruskal(*groups)
        method = "Kruskal-Wallis"
    else:
        stat, p = f_oneway(*groups)
        method = "ANOVA"

    print(f"{method} Test for {numeric_feature} across {attribute}:")
    print(f"Statistic = {stat:.3f}, p-value = {p:.4f}")
    if p < 0.05:
        print("→ Significant difference detected (possible bias). \n")
    else:
        print("→ No significant difference.\n")


# ANOVA Test for Gender on Experience Years
anova_test('Gender', 'Experience_Years')

# ANOVA Test for Race on Experience Years
anova_test('Race', 'Experience_Years')

# ANOVA Test for Age Group on Experience Years
anova_test('Age_Group', 'Experience_Years')

# ANOVA Test for Gender on Screening Score
anova_test('Gender', 'Screening_Score')

# ANOVA Test for Race on Screening Scores
anova_test('Race', 'Screening_Score')

# ANOVA Test for Age Group on Screening Score
anova_test('Age_Group', 'Screening_Score')
```
